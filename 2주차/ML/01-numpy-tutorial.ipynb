## 배열 합치기

* stack (axis 개수가 증가함)
* concatenate (aixs가 유지)

```python
X = np.array([[1,1],[2,2]])
X
```

```python
Y = np.array([[3,3],[4,4]])
Y
```

```python
np.concatenate((X,Y))
```

```python
np.concatenate((X,Y),axis=1)
```

```python
np.stack((X,Y))
```

```python
np.stack((X,Y),axis=1)
```
