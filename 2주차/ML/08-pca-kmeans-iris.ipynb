## 투영

- 투영을 할때 데이터의 feature를 잘 파악하기 위해선, 분산이 큰쪽에서 투영을 해야함.
- 주성분을 바탕으로 분석해야함 -> PCA
- 첫번째 축에 직교하고 남은 분산을 최대해한 보존하는 두번째 축을 찾음.

```python
df = df.drop("Id",axis=1)
df
```

```python
y = df.Species
X = df.drop("Species",axis=1)
```

```python
X.shape, y.shape
```

```python
# from sklearn.processing import StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_s = scaler.fit_transform(X)

X_s
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_p = pca.fit_transform(X_s)

X_p
```

```python
data = pd.DataFrame(X_p,columns =["PCA1","PCA2"])
data
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_l = le.fit_transform(y)
y_l
```

```python
plt.scatter(data['PCA1'],data['PCA2'],c = y_l, s= 50)
plt.show()
```

# 군집화

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 3)
km = km.fit(X_p)
```

```python
y_pred = km.predict(X_p)
y_pred
```

```python
y_l # y_pred와 실제값인 y_l에 대한 value값이 다르다고 틀린것이 아니라, 군집의 형태를 봐야함.
```

```python
plt.scatter(data['PCA1'],data['PCA2'],c = y_pred, s= 50)
plt.show()
```

```python
plt.scatter(data['PCA1'],data['PCA2'],c = y_l, s= 50)
plt.show()
```

# K 값이 모른다 가정, 이때 최적의 k값 찾기

```python
from sklearn.cluster import KMeans
```

```python
inertial= []

for k in range(2,10):
  km = KMeans(n_clusters=k)
  km.fit(X_p)
  inertial.append(km.inertia_)
```

```python
inertial
```

```python
plt.plot(np.arange(2,10),inertial,'--')
plt.show()
```
