```python
test_df = pd.read_csv("/content/HumanActivityRecognition/test.csv")
train_df = pd.read_csv("/content/HumanActivityRecognition/train.csv")
```

```python
replacement_dict = {'STANDING': 0, 'SITTING': 1, 'LAYING': 2,'WALKING':3,'WALKING_DOWNSTAIRS':4,'WALKING_UPSTAIRS':5}
train_df['Activity'] = train_df['Activity'].replace(replacement_dict)
train_df
```

```python
train_df_target = train_df.Activity
train_df_train = train_df.drop('Activity',axis=1)

print(train_df_train.shape, train_df_target.shape)
```

```python
train_df_train = train_df_train.values
train_df_train
```

```python
train_df_target = train_df_target.values
train_df_target
```

```python
from sklearn.svm import SVC

clf = SVC(random_state = 42)
clf = clf.fit(train_df_train,train_df_target)
```

```python
test_df
```

```python
replacement_dict = {'STANDING': 0, 'SITTING': 1, 'LAYING': 2,'WALKING':3,'WALKING_DOWNSTAIRS':4,'WALKING_UPSTAIRS':5}
test_df['Activity'] = test_df['Activity'].replace(replacement_dict)
test_df
```

```python
test_df_target = test_df.Activity
test_df_train = test_df.drop('Activity',axis=1)

print(test_df_train.shape, test_df_target.shape)
```

```python
test_df_train = test_df_train.values
test_df_target = test_df_target.values
```

```python
y_pred = clf.predict(test_df_train)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score,f1_score


acc = accuracy_score(test_df_target,y_pred) # 정답값, 예측

print(f'optimal_accuracy: {acc}')
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "C" : [0.1 ,1 ],
    'gamma' : [0.01,0.1],
    'degree' : [2,3],
    'kernel' : ['linear','poly']
}

grid_cv = GridSearchCV(SVC(),params,cv=5,refit= True, verbose=3)
grid_cv.fit(train_df_train,train_df_target)
```

```python
grid_cv.best_estimator_
```

```python
grid_cv.best_score_   # 최고의 모델 정확도
```

```python
clf = SVC(C= 1, degree=2, gamma = 0.01, kernel = 'linear')
clf = clf.fit(train_df_train,train_df_target)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score,f1_score


acc = accuracy_score(test_df_target,y_pred) # 정답값, 예측

print(f'optimal_accuracy: {acc}')
```
